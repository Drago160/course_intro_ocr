```python
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms as tr
img, label = train_helper.get_item(46)

image_prepare = tr.Compose([
    tr.ToPILImage(),
    tr.Resize(24),
    tr.CenterCrop(32),
    tr.ToTensor(),
])
def prepare_img(img):
    img = image_prepare(img)
    img[(img == 0)] = 1 # change frame color to white
    return img[0]
plt.imshow(prepare_img(img), cmap='gray')
```

```python
class HWDBDataset(Dataset):
    def __init__(self, helper: HWDBDatasetHelper):
        self.helper = helper
    
    def __len__(self):
        return self.helper.size()
    
    def __getitem__(self, idx):
        img, label = self.helper.get_item(idx)
        return prepare_img(img), label
```

```python
train_dataset = HWDBDataset(train_helper)
val_dataset = HWDBDataset(val_helper)

train_loader = DataLoader(train_dataset, batch_size=512, shuffle=True, drop_last=True, num_workers=8)
val_loader = DataLoader(val_dataset, batch_size=2048, shuffle=False, num_workers=8)
```

## Resnet Block из курса по DL:

```python
import torch.nn.functional as F
```

```python
class BasicBlock(nn.Module):
    expansion = 1

    def __init__(self, in_planes, planes, stride=1):
        super().__init__()
        self.conv1 = nn.Conv2d(in_planes, planes, kernel_size=3,
                               stride=stride, padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(planes)
        self.conv2 = nn.Conv2d(planes, planes, kernel_size=3,
                               stride=1, padding=1, bias=False)
        self.bn2 = nn.BatchNorm2d(planes)

        self.shortcut = nn.Sequential()
        # How to do shortcuts?
        if stride != 1 or in_planes != self.expansion*planes:
            self.shortcut = nn.Sequential(
                nn.Conv2d(in_planes, self.expansion*planes,
                          kernel_size=1, stride=stride, bias=False),
                nn.BatchNorm2d(self.expansion*planes)
            )

    def forward(self, x):
        out = F.relu(self.bn1(self.conv1(x)))
        out = self.bn2(self.conv2(out))
        out += self.shortcut(x)
        out = F.relu(out)
        return out
```

```python
class MyResnetLikeModel(nn.Module):
    def __init__(self, num_classes):
        super().__init__()
        self.in_planes = 32
        self.conv = nn.Conv2d(1, 32, kernel_size=3,
                               stride=1, padding=1, bias=False)
        self.bn = nn.BatchNorm2d(32)
        self.skip_layers = nn.Sequential(
            self.make_layer(32, 1),
            self.make_layer(64, 2),
            self.make_layer(128, 2),
            self.make_layer(256, 2),   
        )
        
        self.linear = nn.Linear(256, num_classes)

    def make_layer(self, planes, stride):
        first_block = BasicBlock(self.in_planes, planes, stride)
        self.in_planes = planes
        sec_block = BasicBlock(self.in_planes, planes, 1)
        return nn.Sequential(
            first_block,
            sec_block 
        )

    def forward(self, x):
        out = self.conv(x)
        out = self.bn(out)
        out = F.relu(out)

        out = self.skip_layers(out)
        
        out = F.avg_pool2d(out, 4)
        out = out.view(out.size(0), -1)
        out = self.linear(out)
        return out
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model = MyResnetLikeModel(train_helper.vocabulary.num_classes())
model = model.to(device)
optim = torch.optim.AdamW(model.parameters(), lr=0.001)
loss_fn = torch.nn.CrossEntropyLoss()
```

```python
from tqdm import tqdm


def run_validation(val_loader: DataLoader, model: nn.Module, n_steps=None):
    model.eval()
    n_good = 0
    n_all = 0
    wrapper = lambda x: x
    if n_steps is None:
        n_steps = len(val_loader)
        wrapper = tqdm
    
    with torch.no_grad():
        for batch, (X, y) in enumerate(wrapper(val_loader)):
            if batch == n_steps:
                break
            logits = model(X.unsqueeze(1).to(torch.float32).cuda())
            classes = torch.argmax(logits, dim=1).cpu().numpy()
            n_good += sum(classes == y.cpu().numpy())
            n_all += len(classes)
    
    return n_good / n_all


def train_epoch(train_loader: DataLoader, val_loader: DataLoader, model: nn.Module, optim, loss_fn):
    model.train()
    
    for batch, (X, y) in enumerate(tqdm(train_loader)):
        X = X.to(device) 
        y = y.to(device)
        
        logits = model(X.unsqueeze(1))
        loss = loss_fn(logits, y)
        
        optim.zero_grad()
        loss.backward()
        optim.step()
```

```python
for epoch in range(10):
    print(f'Epoch {epoch}:')
    train_epoch(train_loader, val_loader, model, optim, loss_fn)
    accuracy = run_validation(val_loader, model)
    print(f'accuracy: {accuracy}')
    torch.save(model.state_dict(), f'pth/my_resnet_like_epoch_{epoch}.pth')
```
